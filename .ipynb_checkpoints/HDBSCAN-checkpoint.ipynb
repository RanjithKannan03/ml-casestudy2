```python
import h5py
import numpy as np
import pickle
import hdbscan
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score, v_measure_score
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
pge_path = 'colon_nct_feature/pge_dim_reduced_feature.h5'
resnet50_path = 'colon_nct_feature/resnet50_dim_reduced_feature.h5'
pge_content = h5py.File(pge_path, mode='r')
resnet50_content = h5py.File(resnet50_path, mode='r')
```

```python
pge_pca_feature  = pge_content['pca_feature'][...]
resnet50_pca_feature  = resnet50_content['pca_feature'][...]
pge_umap_feature  = pge_content['umap_feature'][...]
resnet50_umap_feature = resnet50_content['umap_feature'][...]
```

```python
filename  = np.squeeze(pge_content['file_name'])
filename = np.array([str(x) for x in filename])
labels = np.array([x.split('/')[2] for x in filename])
print(f"Number of labels: {len(set(labels))}")
print(f"Labels: {set(labels)}")
```

```python
def calculate_percent(sub_df, attrib):
    cnt = sub_df[attrib].count()
    output_sub_df = sub_df.groupby(attrib).count()
    return (output_sub_df/cnt)
```

## PathologyGAN

### PCA feature

```python
pca_clusterer = hdbscan.HDBSCAN(min_cluster_size=int(pge_pca_feature.shape[0]*0.01),gen_min_span_tree=True)
print(f"Min CLuster Size: {pca_clusterer.min_cluster_size}")
```

```python
pca_clusterer.fit(pge_pca_feature)
plt.figure(figsize=(10, 7))
pca_clusterer.condensed_tree_.plot(select_clusters=True, selection_palette=sns.color_palette())
plt.show()
```

Since the HDBSCAN condensed tree shows that most points merge into only two stable clusters, we need to reduce the min_cluster_size parameter so that the algorithm can detect smaller, more fine-grained clusters instead of collapsing them into large ones.

```python
pca_clusterer = hdbscan.HDBSCAN(min_cluster_size=int(pge_pca_feature.shape[0]*0.0009),gen_min_span_tree=True)
print(f"Min CLuster Size: {pca_clusterer.min_cluster_size}")
```

```python
pca_clusterer.fit(pge_pca_feature)
plt.figure(figsize=(10, 7))
pca_clusterer.condensed_tree_.plot(select_clusters=True, selection_palette=sns.color_palette())
plt.show()
```

```python
pca_hdbscan_assignment = pca_clusterer.labels_
mask=pca_hdbscan_assignment>-1
pca_hdbscan_assignment=pca_hdbscan_assignment[mask]
unique_clusters = sorted(set(pca_hdbscan_assignment))
masked_labels=labels[mask]

print("Number of clusters exclusing noise:", len(unique_clusters))

count={}

for i in pca_hdbscan_assignment:
    if i in count:
        count[i]+=1
    else:
        count[i]=1
```

```python
resulted_cluster_df = pd.DataFrame({'clusterID': pca_hdbscan_assignment, 'type': masked_labels})
label_proportion_df = resulted_cluster_df.groupby(['clusterID']).apply(lambda x: calculate_percent(x,'type')).rename(columns={'clusterID':'type_occurrence_percentage'}).reset_index()
pivoted_label_proportion_df = pd.pivot_table(label_proportion_df, index = 'clusterID', columns = 'type', values = 'type_occurrence_percentage')

number_of_tile_df = resulted_cluster_df.groupby('clusterID')['type'].count().reset_index().rename(columns={'type':'number_of_tile'})
df_idx = pivoted_label_proportion_df.index
(pivoted_label_proportion_df*100).loc[df_idx].plot.bar(stacked=True)

plt.ylabel('Percentage of tissue type')
plt.legend(loc='upper right')
plt.title('Cluster configuration by HDBSCA - PCA excluding noise')

plt.show()
```

```python
v_score_pca=v_measure_score(masked_labels, pca_hdbscan_assignment)

print(f"V-Score for PGE+PCA: {v_score_pca:4f}")
```

```python
print(f"Number of points classified as noise: {len(labels)-len(masked_labels)}")
```

#### Conclusion: Since PCA compresses local structure and reduces density contrast, HDBSCAN on PCA features yields a low V-score. Since the local structures are compressed, this causes HDBSCAN to classify many points as noise because it cannot detect enough dense regions to form stable clusters.

### Conclusion: Quality vs. Coverage Trade-off
HDBSCAN shows a trade-off between cluster purity and coverage:
- High-Precision (min_cluster_size=50): V-Score 0.62, highly accurate clusters, but many points marked as noise (844).
- High-Coverage (min_cluster_size=60): V-Score 0.54, recovers 573 noise points, but clusters are less distinct.

To handle the 844 noise points without sacrificing the 0.62 V-score, apply Soft Clustering

```python
umap_clusterer = hdbscan.HDBSCAN(
    min_cluster_size=int(pge_pca_feature.shape[0]*0.01),
    gen_min_span_tree=True,
    prediction_data=True
)
print(f"Min CLuster Size: {umap_clusterer.min_cluster_size}")
```

```python
umap_clusterer.fit(pge_umap_feature)
plt.figure(figsize=(10, 7))
umap_clusterer.condensed_tree_.plot(select_clusters=True, selection_palette=sns.color_palette())
plt.show()
```

```python
soft_clusters = hdbscan.all_points_membership_vectors(umap_clusterer)
refined_labels = umap_clusterer.labels_.copy()
probability_threshold = 0.5

noise_indices = np.where(refined_labels == -1)[0]
rescued_count=0
```

```python
for i in noise_indices:
    probs = soft_clusters[i]
    if np.max(probs) >= probability_threshold:
        refined_labels[i] = np.argmax(probs)
        rescued_count += 1
```

```python
print(f"Original Noise Count: {len(noise_indices)}")
print(f"Points Rescued via Soft Clustering: {rescued_count}")
print(f"Remaining Noise: {len(noise_indices) - rescued_count}")
```

```python
final_clean_mask = refined_labels>-1
final_clean_labels = refined_labels[final_clean_mask]
unique_clusters = sorted(set(final_clean_labels))
masked_labels=labels[final_clean_mask]

print("Number of clusters exluding noise:", len(unique_clusters))

count={}

for i in final_clean_labels:
    if i in count:
        count[i]+=1
    else:
        count[i]=1
```

```python
v_score_umap=v_measure_score(masked_labels, final_clean_labels)
print(f"V-Score for PGE+UMAP: {v_score_umap:4f}")
```

```python
print(f"DBCV Score: {umap_clusterer.relative_validity_:.3f}")
```

#### Silhouette score is not suitable for HDBSCAN because it assumes all points belong to clusters and favors convex, well-separated clusters, which penalizes HDBSCANs irregular and varying-density clusters. Many points in HDBSCAN are labeled as noise, which silhouette cannot handle properly. DBCV, on the other hand, is designed for density-based clustering, naturally accounts for noise, and evaluates clusters based on density and connectivity, providing a more realistic measure of cluster quality.
