```python
# IMPORTS
import h5py # Required for loading the .h5 files
import numpy as np # Required for numerical computations
import matplotlib.pyplot as plt # Required for plotting graphs
from sklearn.preprocessing import LabelEncoder # Required for encoding string labels to integers
from sklearn.cluster import KMeans # Required for K-Means Clustering
# Import all required metrics: ARI (Extrinsic), Silhouette & V-Measure (Intrinsic)
from sklearn.metrics import adjusted_rand_score, silhouette_score, v_measure_score
from sklearn.neighbors import NearestNeighbors # Required for building the Nearest Neighbors model
import networkx as nx  # Required for building the Nearest Neighbor Graph
import community  # Required for running the Louvain Community Detection algorithm
```

```python
# File paths
PGE_FILE = 'colon_nct_feature/colon_nct_feature/pge_dim_reduced_feature.h5'
RESNET50_FILE = 'colon_nct_feature/colon_nct_feature/resnet50_dim_reduced_feature.h5'
```

```python
# Load data from HDF5 files
try:
    with h5py.File(PGE_FILE, 'r') as pge_content, \
            h5py.File(RESNET50_FILE, 'r') as resnet_content:
        # PGE Features
        file_names = [name.decode('utf-8')
                      for name in pge_content['file_name'][...]]
        pge_pca = pge_content['pca_feature'][...]
        pge_umap = pge_content['umap_feature'][...]

        # ResNet50 Features
        resnet_pca = resnet_content['pca_feature'][...]
        resnet_umap = resnet_content['umap_feature'][...]

except FileNotFoundError:
    print(f"!!! CRITICAL ERROR: Required files not found. !!!")
    print(
        f"and contains these files: '{PGE_FILE}' and '{RESNET50_FILE}' in the current directory.")
    exit()
```

```python
# Extract true labels from file names

labels_raw = [name.split('/')[2] for name in file_names]
label_encoder = LabelEncoder()
true_labels = label_encoder.fit_transform(labels_raw)
class_names = label_encoder.classes_ # 9 Tissue Types

# Confirmation print
print(f"Features loaded. True Labels (K={len(class_names)}) set: {', '.join(class_names)}")
```

```python
# K-Means Optimal K Search using Elbow Method
# Elbow Method Definition: Elbow Method is a technique used to determine the optimal number of clusters (K) in 
# K-Means clustering by plotting the inertia (WCSS) against different K values and identifying the "elbow" point 
# where the rate of decrease sharply changes. This point indicates a balance between cluster compactness and model simplicity. 

def find_optimal_k(features, max_k=15):
    inertia = []
    k_range = range(1, max_k + 1)
    
    for k in k_range:
        # K-Means Model Fitting & Inertia Calculation 
        model = KMeans(n_clusters=k, random_state=42, n_init=10)
        model.fit(features)
        inertia.append(model.inertia_)

    # Graph Plotting
    plt.figure(figsize=(10, 6))
    plt.plot(k_range, inertia, marker='o', linestyle='-', color='red')
    plt.title('Elbow Method: Optimal K (PGE-PCA features)')
    plt.xlabel('Number of Clusters (K)')
    plt.ylabel('Inertia (WCSS)')
    plt.xticks(k_range)
    plt.grid(True)
    plt.show() 
    
    # Analysis based on Elbow:
    K_OPTIMAL_LOCAL = len(class_names) # Based on known tissue classes
    print(f"Analysis: Based on known tissue classes from true labels, setting Optimal K = {K_OPTIMAL_LOCAL}")
    return K_OPTIMAL_LOCAL

# Call the function to find optimal K
K_OPTIMAL = find_optimal_k(pge_pca)
print(f"Optimal K determined from Elbow Method: {K_OPTIMAL}")
```

```python
def run_kmeans_and_evaluate(features, feature_set_name, k_optimal, true_labels):
    """ Runs K-Means clustering and evaluates using ARI, Silhouette, and V-Measure."""
    # Added defensive check to ensure k_optimal is an integer before initializing KMeans
    if not isinstance(k_optimal, int) or k_optimal < 1:
        raise ValueError(
            f"Invalid K value received: {k_optimal}. K must be an integer >= 1.")

    kmeans = KMeans(n_clusters=k_optimal, random_state=42, n_init=10)
    kmeans.fit(features)
    predicted_labels = kmeans.labels_

    # Extrinsic Measures (Require True Labels)
    ari_score = adjusted_rand_score(true_labels, predicted_labels)
    v_measure = v_measure_score(true_labels, predicted_labels)

    # Intrinsic Measure (Does NOT Require True Labels). High compute cost.
    try:
        # Check if enough samples/clusters exist to compute Silhouette score
        if len(np.unique(predicted_labels)) > 1 and len(features) > 1:
            silhouette = silhouette_score(features, predicted_labels)
        else:
            silhouette = np.nan
    except:
        silhouette = np.nan  # Use NaN if computation fails/is too slow

    return {
        'set': feature_set_name,
        'ari': ari_score,
        'silhouette': silhouette,
        'v_measure': v_measure,
        'labels': predicted_labels
    }

# Running K-Means on Different Feature Sets and Collecting Results
all_results = []

# --- FIX: Defensive Check for K_OPTIMAL to prevent InvalidParameterError ---
# If K_OPTIMAL is None (due to skipped execution of preceding cell), run dependency
if 'K_OPTIMAL' not in globals() or K_OPTIMAL is None:
    print("!!! WARNING: K_OPTIMAL was not defined or set. Running dependency (find_optimal_k) now.")
    try:
        # Re-run the dependency calculation using pge_pca
        K_OPTIMAL = find_optimal_k(pge_pca)
    except NameError:
        # Fallback if pge_pca is not even defined
        K_OPTIMAL = 9  # Use the known optimal value as a last resort
        print(
            f"!!! CRITICAL WARNING: Data variables (like pge_pca) not found. Defaulting K_OPTIMAL to {K_OPTIMAL}. Please ensure all preceding steps are run.")


# Run K-Means on Different Feature Sets
all_results.append(run_kmeans_and_evaluate(
    pge_pca, 'PGE + PCA', K_OPTIMAL, true_labels))
all_results.append(run_kmeans_and_evaluate(
    pge_umap, 'PGE + UMAP', K_OPTIMAL, true_labels))
all_results.append(run_kmeans_and_evaluate(
    resnet_pca, 'ResNet50 + PCA', K_OPTIMAL, true_labels))
all_results.append(run_kmeans_and_evaluate(
    resnet_umap, 'ResNet50 + UMAP', K_OPTIMAL, true_labels))

# Displaying Results
print("\n Feature Set | ARI (Extrinsic) | V-Measure | Silhouette (Intrinsic) |")
best_ari_score = -1
best_result_kmeans = None

for res in all_results:
    print(
        f"| {res['set']} | {res['ari']:.4f} | {res['v_measure']:.4f} | {res['silhouette']:.4f} |")
    if res['ari'] > best_ari_score:
        best_ari_score = res['ari']
        best_result_kmeans = res

# Best Feature Set Selection
features_best = resnet_umap
feature_set_name_best = best_result_kmeans['set']
```

```python
K_NEIGHBORS = 7  # Heuristic value for graph construction

# Nearest Neighbors Model Creation
nn_model = NearestNeighbors(n_neighbors=K_NEIGHBORS,
                            metric='euclidean', n_jobs=-1)
nn_model.fit(features_best)
distances, indices = nn_model.kneighbors(features_best)

graph = nx.Graph()
num_samples = features_best.shape[0]
graph.add_nodes_from(range(num_samples))

# Graph Construction (Edges with Weight = Inverse Distance)
for i in range(num_samples):
    for j in range(1, K_NEIGHBORS):
        neighbor_index = indices[i, j]
        distance = distances[i, j]
        weight = 1.0 / (distance + 1e-6)
        graph.add_edge(i, neighbor_index, weight=weight)

# Louvain Community Detection on Graph using True Labels and Louvain Labels as Ground Truth and Predicted Labels respectively (Extrinsic, Intrinsic)
partition = community.best_partition(graph, random_state=42)
louvain_labels = np.array([partition[node] for node in graph.nodes()])

# Louvain Metrics Calculation (ARI, V-Measure, Silhouette) using True Labels and Louvain Labels as Ground Truth and Predicted Labels respectively (Extrinsic, Intrinsic)    
ari_louvain = adjusted_rand_score(true_labels, louvain_labels)
v_measure_louvain = v_measure_score(true_labels, louvain_labels)
try:
    silhouette_louvain = silhouette_score(features_best, louvain_labels)
except:
    silhouette_louvain = np.nan

print("\n Louvain Results (on ResNet50 + UMAP)")
print(f" Metric        | Score ")
print(f" --------------+-------")
print(f" ARI           | {ari_louvain:.4f} ")
print(f" V-Measure     | {v_measure_louvain:.4f} ")
print(f" Silhouette    | {silhouette_louvain:.4f} ")
```

```python
# Final Visualization of Best K-Means Clustering Results
predicted_labels = best_result_kmeans['labels']

fig, axes = plt.subplots(1, 2, figsize=(18, 7))

# Plot 1: True Labels (Ground Truth) on Best Feature Set (ResNet50 + UMAP) 
scatter1 = axes[0].scatter(
    features_best[:, 0], features_best[:, 1], c=true_labels, cmap='Spectral', s=20)
axes[0].set_title(
    f'1. Ground Truth (True Labels) on {feature_set_name_best}', fontsize=14)
axes[0].set_xlabel('UMAP Dimension 1')
axes[0].set_ylabel('UMAP Dimension 2')
legend1 = axes[0].legend(*scatter1.legend_elements(),
                         title="Tissue Types", labels=list(class_names), loc="lower left")
axes[0].add_artist(legend1)

# Plot 2: Predicted K-Means Clusters on Best Feature Set (ResNet50 + UMAP)
scatter2 = axes[1].scatter(
    features_best[:, 0], features_best[:, 1], c=predicted_labels, cmap='Spectral', s=20)
axes[1].set_title(
    f'2. Predicted K-Means Clusters (K={K_OPTIMAL}) on {feature_set_name_best}', fontsize=14)
axes[1].set_xlabel('UMAP Dimension 1')
axes[1].set_ylabel('UMAP Dimension 2')
legend2 = axes[1].legend(*scatter2.legend_elements(),
                         title="Cluster ID", loc="lower left")
axes[1].add_artist(legend2)

plt.tight_layout()
plt.show()
```
